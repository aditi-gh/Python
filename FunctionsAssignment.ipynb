1. What is the difference between a function and a method in Python?
- Function
  - A block of code that performs a specific task.
  - Defined using def keyword.
  - Can exist independently (not tied to any object).
  - Called directly by its name.
  - Example of a Function
    - def add(a, b):
    - return a + b
    - print(add(5, 3))   # Output: 8
- Method
  - A function that belongs to an object (inside a class).
  - Defined like a function but always takes self (the instance) as its first parameter.
  - Called on an object using dot (.) notation.
  - Example of a Method
    - class Calculator:
      - def multiply(self, a, b):   # 'self' refers to the object
        - return a * b
    - calc = Calculator()
    - print(calc.multiply(4, 2))   # Output: 8

2. Explain the concept of function arguments and parameters in Python.
- Parameters
  - Variables listed inside the parentheses when defining a function.
  - They act as placeholders for values.
  - Example: def greet(name): → here name is a parameter.
- Arguments
  - The actual values you pass into the function when calling it.
  - Example: greet("Aditi") → here "Aditi" is an argument.
- Example
  - def greet(name):     # 'name' is a parameter
      - print("Hello,", name)
  - greet("Aditi")       # "Aditi" is an argument
  - Output: Hello, Aditi


```python
# 3. What are the different ways to define and call a function in Python?

#Simple Function (using def)
def greet():
    print("Hello, World!")

greet()

#Function with Parameters
def add(a, b):
    return a + b

print(add(5, 3))   # Calling with arguments → Output: 8

#Function with Default Parameters
def greet(name="Guest"):
    print("Hello,", name)
greet()          # Uses default → Hello, Guest
greet("Aditi")   # Overrides default → Hello, Aditi

#Function with Variable-Length Arguments
'''
*args → multiple positional arguments
**kwargs → multiple keyword arguments
'''
def show_numbers(*args):
    print(args)
show_numbers(1, 2, 3)   # Output: (1, 2, 3)
def show_info(**kwargs):
    print(kwargs)
show_info(name="Aditi", age=20)   # Output: {'name': 'Aditi', 'age': 20}

#Anonymous Function (Lambda Function)
square = lambda x: x * x
print(square(5))   # Output: 25

#Nested Function (Function inside a Function)
def outer():
    def inner():
        return "Inner function called"
    return inner()
print(outer())   # Output: Inner function called

#Recursive Function (Function calling itself)
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)
print(factorial(5))   # Output: 120

#Function as an Object (Assigning to a variable)
def greet(name):
    return f"Hello, {name}"
say_hello = greet   # Function assigned to variable
print(say_hello("Mou"))   # Output: Hello, Mou
```

4. What is the purpose of the 'return' statement in a Python function?
- The return statement is used inside a function to:
  - Send a value back to the caller (the place where the function was called).
  - Exit the function immediately once it is executed.
  - Allow functions to be used in expressions or assignments (not just print values).

5. What are iterators in Python and how do they differ from iterables?
- Iterable
  - An object that can be looped over.
  - Examples: lists, tuples, strings, sets, dictionaries.
  - It implements the __iter__() method which returns an iterator.
  - We can think of an iterable as a collection of items (like a book).
- Iterator
  - An object that produces values one at a time when we iterate over it.
  - It implements two methods:
    - __iter__() → returns the iterator object itself.
    - __next__() → returns the next item, and raises StopIteration when items are finished.
  - Think of an iterator as a bookmark that remembers where you are in the book.

6.  Explain the concept of generators in Python and how they are defined.
- Generators are special iterators that let you iterate through a sequence of values lazily (one at a time, on demand). Instead of computing and storing all values at once (like lists), they generate values only when needed. This makes them memory-efficient for handling large datasets.
- There are two ways to define generators in Python:
  - Using a Function with yield
    - Similar to a normal function, but uses yield instead of return.
    - Each time yield is executed, the function pauses and resumes from there on the next call.
  - Using Generator Expressions (like list comprehensions but with parentheses)
    - squares = (x*x for x in range(5))
    - print(next(squares))  # 0
    - print(next(squares))  # 1
    - print(next(squares))  # 4

7. What are the advantages of using generators over regular functions?
- Advantages of Generators over Regular Functions:
  - Memory Efficient - Regular functions with return produce the entire result at once, storing everything in memory.
  - Faster for Large Data - Since values are generated on the fly, generators are faster when handling large streams of data (like files, logs, infinite sequences).
  - Infinite Sequences Possible - A regular function must return all values (impossible for infinite data). A generator can yield values forever.
  - Pipeline Processing - Generators can be chained together, allowing stream processing (like Unix pipes).
  - Cleaner Code - No need to manage __iter__() and __next__() methods manually. Just write a function with yield and Python builds the iterator for you.

8. What is a lambda function in Python and when is it typically used?
- A lambda function is a small, anonymous (nameless) function defined using the lambda keyword.
  - It can take any number of arguments but has only one expression.
  - The expression is automatically returned.
- Lambda functions are typically used when:
  - You need a short throwaway function (instead of writing a full def).
  - Used as arguments to higher-order functions like map(), filter(), reduce().

9. Explain the purpose and usage of the 'map()' function in Python.
- The map() function is used to apply a function to each element of an iterable (like a list, tuple, set, etc.) and returns a map object (which is an iterator).
  - General syntax: map(function, iterable)
  - function → the function you want to apply (can be normal or lambda).
  - iterable → sequence of data (list, tuple, etc.).
- Example 1: Using with a Built-in Function
  - numbers = [1, 2, 3, 4]
  - result = map(str, numbers)   #Convert each number to string
print(list(result))  # ['1', '2', '3', '4']


10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- map()
  - Applies a function to each element of an iterable.
  - Input → iterable
  - Output → transformed iterable (same length)
- filter()
  - Filters elements from an iterable based on a condition (function must return True/False).
  - Input → iterable
  - Output → subset of original iterable
- reduce()
  - Repeatedly applies a function to the iterable to reduce it to a single value.
  - Input → iterable
  - Output → single value
  - Must be imported from functools.

```python
# Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
  def __init__(self, num_terms):
    self.num_terms = num_terms
    self.current_term = 0
    self.a = 0
    self.b = 1

  def __iter__(self):
    return self

  def __next__(self):
    if self.current_term < self.num_terms:
      if self.current_term == 0:
        self.current_term += 1
        return self.a
      elif self.current_term == 1:
        self.current_term += 1
        return self.b
      else:
        next_fib = self.a + self.b
        self.a = self.b
        self.b = next_fib
        self.current_term += 1
        return next_fib
    else:
      raise StopIteration

# Example usage:
fib_iterator = FibonacciIterator(10)

print("Fibonacci sequence up to 10 terms:")
for term in fib_iterator:
  print(term)
```

```python
#Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(exponent):
  for i in range(exponent + 1):
    yield 2 ** i
# Example usage:
power_generator = powers_of_two(5)

print("Powers of 2 up to exponent 5:")
for power in power_generator:
  print(power)
```

```python
# Implement a generator function that reads a file line by line and yields each line as a string.
def file_line_generator(filepath):
  try:
    with open(filepath, 'r') as f:
      for line in f:
        yield line.strip()
  except FileNotFoundError:
    print(f"Error: File not found at {filepath}")
  except Exception as e:
    print(f"An error occurred: {e}")

# Example usage:
file_path = "my_dummy_file.txt"
line_generator = file_line_generator(file_path)

print(f"Reading lines from '{file_path}':")
for line in line_generator:
  print(line)
```

```python
# Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
  return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temperatures = [0, 10, 20, 30, 40, 100]

# Use map() to apply the conversion function to each element in the list
fahrenheit_temperatures_map_object = map(celsius_to_fahrenheit, celsius_temperatures)

# Convert the map object to a list to display the results
fahrenheit_temperatures = list(fahrenheit_temperatures_map_object)

print(f"Celsius temperatures: {celsius_temperatures}")
print(f"Fahrenheit temperatures: {fahrenheit_temperatures}")
```

```python
#Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(input_string):
  vowels = 'aeiouAEIOU'
  filtered_string = ''.join(filter(lambda x: x not in vowels, input_string))
  return filtered_string
```
